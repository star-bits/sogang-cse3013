```python
def parse_maze(maze_str):
    lines = maze_str.strip().split('\n')
    height = (len(lines) - 1) // 2
    width = (len(lines[0]) - 1) // 2
    graph = {i: [] for i in range(height * width)}

    for i in range(height):
        for j in range(width):
            cell_index = i * width + j
            # Check right neighbor
            if j < width - 1 and lines[2 * i + 1][2 * j + 2] == ' ':
                graph[cell_index].append(cell_index + 1)
                graph[cell_index + 1].append(cell_index)
            # Check bottom neighbor
            if i < height - 1 and lines[2 * i + 2][2 * j + 1] == ' ':
                graph[cell_index].append(cell_index + width)
                graph[cell_index + width].append(cell_index)

    return graph

# Example usage
maze = """
+-+-+-+-+-+
| | |   | |
+ + + +-+ +
| |   | | |
+ + +-+ + +
|   | |   |
+ +-+ +-+ +
| | |   | |
+ + +-+ + +
|         |
+-+-+-+-+-+
"""

graph = parse_maze(maze)
for key, value in graph.items():
    print(f"{key}: {value}")
```
